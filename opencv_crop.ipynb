```python
methods = [
    ("INTER_NEAREST", cv2.INTER_NEAREST),
    ("INTER_LINEAR", cv2.INTER_LINEAR),
    ("INTER_AREA", cv2.INTER_AREA),
    ("INTER_CUBIC", cv2.INTER_CUBIC),
    ("INTER_LANCZOS4", cv2.INTER_LANCZOS4)]
```

```python
for method, tech in methods:
    print(f"{method}: {tech}")
    resized = imutils.resize(boat,
                             boat.shape[1] * 5,
                             inter=tech)
    cv2.imwrite(os.path.join(path_folder, f"{method}.jpg"), resized)
    display(Image(filename=os.path.join(path_folder, f"{method}.jpg")))
```

```python
# Boucle pour rotation d'image par incréments de 45 degrés
angles = range(0, 360, 45)  # De 0 à 360 degrés par pas de 45

for i, angle in enumerate(angles):
    # Rotation de l'image avec imutils
    rotated = imutils.rotate(image, angle=angle)
    
    # Sauvegarde de l'image rotée
    filename = f"rotation_{angle}deg.jpg"
    cv2.imwrite(os.path.join(path_folder, filename), rotated)
    
    # Affichage de l'image
    print(f"Rotation {angle}°")
    display(Image(filename=os.path.join(path_folder, filename)))
```

```python
angle2 = range(0, 360, 15)

for i, angle in enumerate(angle2):
    rotated = imutils.rotate(image, angle=angle)
    filename = f"rotation_{angle}deg.jpg"
    cv2.imwrite(os.path.join(path_folder, filename), rotated)
    display(Image(filename=os.path.join(path_folder, filename)))
```
