```python
import cv2
from IPython.display import Image, display
import numpy as np
import os

from numpy.ma.core import shape
```

```python
image = cv2.imread("/Users/leodidier/Documents/GitHub/image_loader_opencv/Saint_Joseph_charpentier.jpg")
```

```python
image_path = "/Users/leodidier/Documents/GitHub/image_loader_opencv/opencv_masking"
```

```python
cv2.imwrite(os.path.join(image_path, "Saint_Joseph_charpentier_mask.jpg"), image)
```

```python
display(Image(filename=os.path.join(image_path, "Saint_Joseph_charpentier_mask.jpg")))
```

```python
mask = np.zeros(image.shape[:2], dtype="uint8")
```

```python
cv2.rectangle(mask, (397, 172), (512, 289), 255, -1) # X,Y, X,Y
```

```python
cv2.imwrite(os.path.join(image_path, "Saint_Joseph_charpentier_mask_rectangle.jpg"), mask)
```

```python
display(Image(filename=os.path.join(image_path, "Saint_Joseph_charpentier_mask_rectangle.jpg")))
```

```python
masked = cv2.bitwise_and(image, image, mask=mask)
```

```python
cv2.imwrite(os.path.join(image_path, "Saint_Joseph_charpentier_mask_rectangle_masked.jpg"), masked)
```

```python
display(Image(filename=os.path.join(image_path, "Saint_Joseph_charpentier_mask_rectangle_masked.jpg")))
```

```python
for i in range(0, 255, 249):
    mask = np.zeros(image.shape[:2], dtype="uint8")
    cv2.rectangle(mask, (397, 172), (512, 289), i, -1)
    masked = cv2.bitwise_and(image, image, mask=mask)
    cv2.imwrite(os.path.join(image_path, "Saint_Joseph_charpentier_mask_rectangle_masked_{}.jpg".format(i)), masked)
    display(Image(filename=os.path.join(image_path, "Saint_Joseph_charpentier_mask_rectangle_masked_{}.jpg".format(i))))
```

```python
for i in range (0, 255, 249):
    mask = np.zeros(image.shape[:2], dtype="uint8")
    cv2.circle(mask, (447, 233), 75, i, -1)
    masked = cv2.bitwise_and(image, image, mask=mask)
    cv2.imwrite(os.path.join(image_path, "Saint_Joseph_charpentier_mask_circle_masked_{}.jpg".format(i)), masked)
    display(Image(filename=os.path.join(image_path, "Saint_Joseph_charpentier_mask_circle_masked_{}.jpg".format(i))))
```
