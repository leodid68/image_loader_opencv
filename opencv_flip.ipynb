```python
import cv2
from IPython.display import Image, display
import numpy as np
import os
```

```python
image_path = "/Users/leodidier/Documents/GitHub/image_loader_opencv/open_cv_flip"
```

```python
image = cv2.imread("Saint_Joseph_charpentier.jpg")
cv2.imwrite(os.path.join(image_path, "Saint_Joseph_charpentier.jpg"), image)
```

```python
display(Image(filename=os.path.join(image_path, "Saint_Joseph_charpentier.jpg")))
```

```python
print("Flipping image horizontally")
flipped_image = cv2.flip(image, 1)
cv2.imwrite(os.path.join(image_path, "flipped_image.jpg"), flipped_image)
display(Image(filename=os.path.join(image_path, "flipped_image.jpg")))
```

```python
image_ok = cv2.imread(os.path.join(image_path, "Saint_Joseph_charpentier.jpg"))
```

```python
sens = [
    ("Flipping image horizontally", 1),
    ("Flipping image vertically", 0),
    ("Flipping image both horizontally and vertically", -1)
]
```

```python
for description, flipcode in sens:
    print(f"{description}: flipcode = {flipcode}")

    flipped_image = cv2.flip(image_ok, flipcode)

    filename = f"Flipped_image_{description}.jpg"
    cv2.imwrite(os.path.join(image_path, filename), flipped_image)
    display(Image(filename=os.path.join(image_path, filename)))
```
